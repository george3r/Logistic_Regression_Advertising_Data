```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ad_data = pd.read_csv('advertising.csv')
```

```python
ad_data
```

```python
ad_data['Timestamp'] = pd.to_datetime(ad_data['Timestamp'])
```

```python
ad_data['Year'] = pd.DatetimeIndex(ad_data['Timestamp']).year
ad_data['Month'] = pd.DatetimeIndex(ad_data['Timestamp']).month
ad_data['Day'] = pd.DatetimeIndex(ad_data['Timestamp']).day
ad_data['Time'] = pd.DatetimeIndex(ad_data['Timestamp']).time
```

```python
ad_data.drop(['Ad Topic Line','City','Country','Timestamp','Time'],axis=1,inplace=True)
ad_data2 = ad_data.drop(['Month','Day'],axis=1) # for analysis without Day and month
```

```python
ad_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data.drop(['Clicked on Ad'],axis=1)
y = ad_data['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
y_train
```

```python
# Logistic Regression WITH Day and Month as variables
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter=300)
logmodel.fit(X_train,y_train)
```

```python
coef_df = pd.DataFrame(logmodel.coef_.reshape(7,1),X_train.columns,columns=['Coef'])
coef_df
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print()
print(confusion_matrix(y_test,predictions))
```
